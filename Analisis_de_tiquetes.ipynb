```python
# Importar las librerías necesarias
import pandas as pd
from google.colab import files

# Función para cargar un archivo de Excel y crear un DataFrame
def cargar_y_auditar_excel():
    # Solicitar al usuario que cargue un archivo
    uploaded = files.upload()
    if uploaded:
        # Leer el primer archivo subido (nota: esto procesará el primer archivo si se suben varios)
        nombre_archivo = next(iter(uploaded))
        df = pd.read_excel(nombre_archivo)

        # Crear un informe de auditoría para el DataFrame
        auditoria = {
            'Nombre de Columna': df.columns,
            'Número de Columna': range(1, len(df.columns) + 1),
            'Tipos de Datos': df.dtypes,
            'Valores Únicos': df.nunique(),
            'Valores Faltantes': df.isnull().sum(),
            'Número Total de Filas': [len(df)] * len(df.columns)  # Repite el total de filas para cada columna
        }
        df = pd.DataFrame(auditoria)

        # Retornar el DataFrame de auditoría
        return df

# Llamar a la función y mostrar el resultado
df = cargar_y_auditar_excel()
print(df)
```

```python
import pandas as pd
import re

# Supongamos que df es tu DataFrame ya cargado

# Función para limpiar la columna "PASAJERO"
def limpiar_pasajero(df):
    # Reemplazar "/" por un espacio
    df['PASAJERO'] = df['PASAJERO'].str.replace("/", " ", regex=True)
    # Quitar signos de puntuación restantes, excepto espacios y caracteres alfanuméricos
    df['PASAJERO'] = df['PASAJERO'].str.replace(r"[^\w\s]", "", regex=True)
    # Quitar espacios sobrantes al inicio y al final
    df['PASAJERO'] = df['PASAJERO'].str.strip()
    # Convertir todo a mayúsculas
    df['PASAJERO'] = df['PASAJERO'].str.upper()
    #QUITAR MR Y MRS
    df['PASAJERO'] = df['PASAJERO'].str.replace(r"\b(MR|MRS)\b", "", regex=True, flags=re.IGNORECASE)

# Llamar a la función con el DataFrame
limpiar_pasajero(df)

# Ver el resultado
print(df['PASAJERO'].head())
```

```python
import pandas as pd

# Asumir que df es tu DataFrame y tiene una columna 'VICEPRESIDENCIA'

# Calcular el número de apariciones de cada valor en 'VICEPRESIDENCIA'
count_series = df['VICEPRESIDENCIA'].value_counts()

# Mapear estos conteos de vuelta al DataFrame original
df['COUNT'] = df['VICEPRESIDENCIA'].map(count_series)

# Ahora que tenemos la columna 'COUNT', podemos aplicar la función de normalización
def normalizar_vicepresidencia(row):
    vp = row['VICEPRESIDENCIA']
    count = row['COUNT']

    # Normalizar todas las instancias de PRESIDENCIA
    if "PRESIDENCIA" in vp:
        return "PRESIDENCIA"

    # Reasignar VICEPRESIDENCIA DE PROMOCION Y FOMENTO a VICEPRESIDENCIA DE PROMOCION Y FOMENTO MINERO
    elif vp == "VICEPRESIDENCIA DE PROMOCION Y FOMENTO":
        return "VICEPRESIDENCIA DE PROMOCION Y FOMENTO MINERO"

    # Mantener nombres específicos que no deben cambiar
    elif vp in ["VICEPRESIDENCIA ADMINISTRATIVA Y FINANCIERA",
                "VICEPRESIDENCIA DE CONTRATACION Y TITULACIÓN MINERA",
                "VICEPRESIDENCIA DE PROMOCION Y FOMENTO MINERO",
                "VICEPRESIDENCIA DE SEGUIMIENTO, CONTROL Y SEGURIDAD MINERA",
                "GRUPO DE SALVAMENTO MINERO", "GRUPO FOMENTO"]:
        return vp

    # Categorías con menos de 50 apariciones a "OTRAS", excepto "GRUPO FOMENTO"
    elif count < 50 and vp != "GRUPO FOMENTO":
        return "OTRAS"

    # Cualquier otro caso no especificado se mantiene
    return vp

# Aplicar la función a cada fila
df['VICEPRESIDENCIA_NORMALIZADA'] = df.apply(normalizar_vicepresidencia, axis=1)

# Mostrar los resultados
print(df[['VICEPRESIDENCIA', 'VICEPRESIDENCIA_NORMALIZADA']])
```

```python
import pandas as pd

# Asumiendo que df es tu DataFrame con 'VICEPRESIDENCIA' y 'COUNT' ya creada

def normalizar_vicepresidencia(row):
    vp = row['VICEPRESIDENCIA'].strip()  # Aseguramos quitar espacios extras
    count = row['COUNT']

    # Reasignar específicamente la VICEPRESIDENCIA DE PROMOCION Y FOMENTO
    if vp == "VICEPRESIDENCIA DE PROMOCION Y FOMENTO" and count == 1:
        return "VICEPRESIDENCIA DE PROMOCION Y FOMENTO MINERO"

    # Mantener nombres específicos que no deben cambiar
    elif vp in [
        "VICEPRESIDENCIA ADMINISTRATIVA Y FINANCIERA",
        "VICEPRESIDENCIA DE CONTRATACION Y TITULACIÓN MINERA",
        "VICEPRESIDENCIA DE PROMOCION Y FOMENTO MINERO",
        "VICEPRESIDENCIA DE SEGUIMIENTO, CONTROL Y SEGURIDAD MINERA",
        "GRUPO DE SALVAMENTO MINERO",
        "GRUPO FOMENTO"
    ]:
        return vp

    # Normalizar todas las instancias de PRESIDENCIA
    elif "PRESIDENCIA" in vp:
        return "PRESIDENCIA"

    # Categorías con menos de 50 apariciones a "OTRAS"
    elif count < 50:
        return "OTRAS"

    # Cualquier otro caso no especificado se mantiene
    return vp

# Aplicar la función a cada fila
df['VICEPRESIDENCIA_NORMALIZADA'] = df.apply(normalizar_vicepresidencia, axis=1)

# Mostrar los valores únicos de la columna normalizada para verificación
print(df['VICEPRESIDENCIA_NORMALIZADA'].unique())
```
